## Data Generation examples

To sample the same data they used in the two toy examples you can just call the generate regression function

```python
y_hsc, X_hsc = generate_regression_outputs(type='hsc')
y_bm, X_bm = generate_regression_outputs(type='bimodal')
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,4))
ax[0].scatter(X_hsc, y_hsc, s=3)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].scatter(X_bm, y_bm, s=3)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
plt.show()
```

You can also specify any given X data set and apply the same transformation functions used in the paper to get different effects

```python
X_test = np.random.normal(0,1, size=1000)
y_test_hsc, _ = generate_regression_outputs(type='hsc', X=X_test)
y_test_bm, _ = generate_regression_outputs(type='bimodal', X=X_test)
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,4))
ax[0].scatter(X_test, y_test_hsc, s=3)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].scatter(X_test, y_test_bm, s=3)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
plt.show()
```
